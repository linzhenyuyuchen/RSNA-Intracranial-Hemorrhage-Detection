```python
res = train2[train2['ID'].isin(test1_ID)]
res.sort_values(by='ID')
```

```python
mix = pd.read_csv('./rsna/mix/se_resnet50_submission_fold1_epoch2tta.csv')
mix.sort_values(by='ID')
```

```python
merge_12 = pd.merge(mix,res,left_on='ID',right_on='ID').reset_index()

merge_12
```

```python
sum = 0

for row in merge_12.iterrows():
    tmp = row[1]['Label_x'] - row[1]['Label_y']
    sum += (tmp)
sum
```

```python
sum = 0

for row in merge_12.iterrows():
    tmp = row[1]['Label_x'] - row[1]['Label_y']
    sum += abs(tmp)
sum
```
